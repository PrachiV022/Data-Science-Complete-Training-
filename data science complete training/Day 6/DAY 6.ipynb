```python
class Remote:
    def __init__(self): #class constructor, magic method, dundle method     #double underscore
        #initialization
        print('object has been created with class Remote')
        print('Now you can use it!!')

        
    def volume_up(self):
        print('volume up by 1 unit')

    def volume_down(self):
        print('volume down by 1 unit')

    def power_on(self):
        print('TV is on')

    def power_off(self):
        print('TV is off')

    def auto_suggest_channel(self):
        import time
        print('searching trending channel')
        time.sleep(3)
        print('youtube! IITM')
    def __del__(self):# class destructor      # double underscore
        print('object has been deleted successfully ')
```

```python
class ATM:
    def __init__(self):
        print('object created')
        self.pin = 1111 #public instance
        self.balance = 2000 #public instance

    def change_pin(self): #public instance
        self.pin = input('enter nnew PIN: ')
        print('Your PIN has been changed!!')

    def show_balance(self):
        print(f''' your current balance is Rs {self.balance}''')

    def atm_performance_status(self):
        print('ATM working fine 100%')
        
```

```python
icici = ATM()
```

```python
icici.show_balance()
```

```python
icici.balance
```

```python
class ATM_version_1:
    def __init__(self):
        print('object created')
        self._pin = 1111 #protected variable
        self._balance = 2000 #protected variable

    def change_pin(self):
        self._pin = input('enter nnew PIN: ')
        print('Your PIN has been changed!!')

    def show_balance(self):
        print(f''' your current balance is Rs {self._balance}''')

    def atm_performance_status(self):
        print('ATM working fine 100%')
        
```

```python
axis = ATM_version_1()
```

```python
axis._pin
```

```python
class ATM_version_2:
    def __init__(self):
        print('object created')
        self.__pin = 1111 #protected variable
        self.__balance = 2000 #protected variable

    def change_pin(self):#public
        self.__pin = input('enter nnew PIN: ')
        print('Your PIN has been changed!!')

    def show_balance(self):
        print(f''' your current balance is Rs {self.__balance}''')

    def __atm_performance_status(self):# private variable(double underscore)
        print('ATM working fine 100%')
```

```python
kotak = ATM_version_2()
```

```python
kotak._ATM_version_2__pin    # single underscore followed by class name to call the variable
```
